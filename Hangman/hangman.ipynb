# HANGMAN

```python
from random_words import RandomWords
import time

# Primeiramente, o jogador deverá escolher a dificuldade do jogo

difficulty = input("""
WELCOME TO HANGMAN!

Please, select the level: 
level 1 (easy)
Level 2 (normal) 
level 3 (expert)
""")

# A dificuldade do jogo será baseada na quantidade de letras de cada palavra

if difficulty == '1':
    length = 3
elif difficulty == '2':
    length = 5
else:
    length = 7
print('''
You have 10 chances. Good luck!''')

time.sleep(1)

# Usando o time.sleep para o codigo demorar 3 segundos para rodar
print('''

Choosing word...

        ''')

time.sleep(3)

print('Go!')

# chosen_word é a variável que ira receber a palavra escolhida aleatoriamente pelo RandomWords
chosen_word = ""
rw = RandomWords()
word = rw.random_word()
while len(word) != length: # Se a quantidade de letras for diferente da escolhida pelo usuário, o código continuará rodando
    rw = RandomWords()
    word = rw.random_word()

else:
    chosen_word += word #quando a quantidade de letras for a mesma, adicionará à variável chosen_word

digitadas = []
acertos = []
erros = 0

while True:
    letter = ""
    for x in chosen_word:
        if x in acertos:
            letter += x
        else:
            letter += " _ "
    print(letter)
    if letter == chosen_word:
        print("You win!")
        break
    attempt = input("\nEnter with a letter:").lower()
    if attempt in digitadas:
        print("You already tried this letter!")
    else:
        digitadas += attempt
        if attempt in chosen_word:
            acertos += attempt
        else:
            
            if attempt not in chosen_word:
                erros += 1
                print("Wrong!")
                if erros < 10:
                    x = 10 - erros
                    print(f"you have {x} more chances")
                else:
                    print('''You lose! 
The word was...''')
                    time.sleep(1)
                    print(chosen_word)
                    break
```
